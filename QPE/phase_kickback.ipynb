Phase Kickback practice in preparation for Quantum Phase Estimator.

Definition - applying control target system (similar to CNOT conceptually) but instead of the target being affected, the control is. Hense the kickback
Used in Phase estimation

CU gate applied on the v qubit is the target and the control is on the plus when referring to diagram in the notes.

```python
from qiskit import ClassicalRegister, QuantumCircuit, QuantumRegister
import numpy as np
#two qubit state for phase kickback
Control_plus = QuantumRegister(1, "Control_plus")
Target_v = QuantumRegister(1, "Target_v")
c = ClassicalRegister(1, "c") #classical register to store measurement(only measuring 1 so only 1 c bit)
qc = QuantumCircuit(Control_plus, Target_v,c)
qc.h(Control_plus) #put control in superposition
qc.x(Target_v) #x puts qubit into 1 state, which is the eigenstate of U where we chose (Z)
qc.cz(Control_plus, Target_v) #if control is 1, apply Z to target
qc.h(Control_plus) #make measurable
qc.measure(Control_plus, c) #measure control qubit
qc.draw('mpl')
```
